```python
# 2)Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)and make data frame.
```

```python
# Downloading imdb top 100 movie's data
URL = ('https://www.imdb.com/list/ls091520106/')
response = requests.get(URL)
soup = BeautifulSoup(response.text, 'lxml')

Names=[]
Ratings=[]
Year_of_release=[]
for i in soup.find_all('h3',class_="lister-item-header"):
    Names.append(i.text.replace('\n',''))
for i in soup.find_all('div',class_="ipl-rating-star small"):
    Ratings.append(i.text.replace('\n',''))
for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    Year_of_release.append(i.text)
```

```python
import pandas as pd
df=pd.DataFrame({'Names':Names,'Ratings':Ratings,'Year_of_release':Year_of_release})
df
```

```python
# 5)a)Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
```

```python
url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
Teams=[]
Matches=[]
Ratings=[]
for i in soup.find_all('span',class_='u-hide-phablet')[0:10]:
    Teams.append(i.text)
Match1=soup.find('td',class_="rankings-block__banner--matches")
Match1
Point1=soup.find('td',class_="rankings-block__banner--points")
Point1

for i in soup.find_all('td',class_="table-body__cell u-center-text")[0:18]:
    Matches.append(i.text)
Total_Matches= [Match1.text] + Matches[::2]
Total_Points=[Point1.text] + Matches[1::2] 
Rating1=soup.find('td',class_="rankings-block__banner--rating u-text-right")
Rating1

for i in soup.find_all('td',class_="table-body__cell u-text-right rating")[0:9]:
    Ratings.append(i.text)
Final_Ratings=[Rating1.text.strip()]+Ratings
 
```

```python
import pandas as pd
df=pd.DataFrame({'Teams':Teams,'Total_Matches':Total_Matches,'Total_Points':Total_Points,'Final_Ratings':Final_Ratings})
df
```

```python
# 7)Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.
```

```python
url='https://coreyms.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
headings=[]
dates=[]
content_all=[]
links=[]
for i in soup.find_all('a',class_="entry-title-link"):
    headings.append(i.text)
for i in soup.find_all('time',class_="entry-time")[0:10]:
    dates.append(i.text)
for i in soup.find_all('div',class_="entry-content")[0:10]:
    content_all.append(i.text)
for i in soup.find_all('iframe',class_="youtube-player"):
    links.append(i['src'])
```

```python
#9) Write a python program to scrape mentioned details from dineout.co.in :i)Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL
```

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
titles=[]
location=[] 
price=[] 
images=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)    
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text.replace('₹',''))
for i in soup.find_all('img',class_="no-img"):
    images.append(i['data-src'])


    
```

```python
#Making dataframe
import pandas as pd
df=pd.DataFrame({'Titles':titles,'Location':location,'Price':price,'Images_url':images})
df
```
