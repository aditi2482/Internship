```python
driver=webdriver.Chrome("chromedriver.exe")
time.sleep(2)
```

```python
driver.get('https://www.naukri.com/')
```

```python
search_field_designation=driver.find_element_by_class_name("suggestor-input")
search_field_designation.send_keys("Data Analyst")
```

```python
search_field_location=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input')#job search bar
search_field_location.send_keys("Banglore")
```

```python
search_button=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]')
search_button.click()
```

```python
job_titles=[]
company_names=[]
locations_list=[]
experience_list=[]
```

```python
titles_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")#locating web element of title
for i in titles_tags:     
    title=i.text        
    job_titles.append(title)  
job_titles[0:10]  
```

```python
companies_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")  # Locating web element of comapny name
for i in companies_tags:     
    company_name=i.text          
    company_names.append(company_name)
company_names[0:10] 
```

```python
experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span")
for i in experience_tags:     
    experience=i.text     
    experience_list.append(experience)
experience_list[0:10]
```

```python
locations_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
for i in locations_tags:     
    location=i.text        
    locations_list.append(location)
locations_list[0:10] 
```

```python
print(len(job_titles),len(company_names),len(experience_list),len(locations_list))
```

```python
jobs=pd.DataFrame({})
jobs['title']=job_titles
jobs['company']=company_names
jobs['experience_required']=experience_list
jobs['location']=locations_list
```

```python
jobs
```

```python
search_button=driver.find_element_by_xpath("/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input")
search_button.click()
```

```python
Company_Name=[]
Salary_Records=[]
Average_salary=[]
Minmax_Salary=[]
Minimum_salary=[]
Maximum_salary=[]
Experience_required=[]
```

```python
company_tags=driver.find_elements_by_xpath("//div[@class='name']/a")
for i in company_tags:
    company=i.text
    Company_Name.append(company)
Company_Name[0:10]
```

```python
salary_tags=driver.find_elements_by_xpath("//div[@class='name']/span")
for i in salary_tags:
    ts=i.text
    Salary_Records.append(ts)
Salary_Records[0:10]
```

```python
minmax_tags=driver.find_elements_by_xpath("//div[@class='value body-medium']")
for i in minmax_tags:
    ms=i.text
    Minmax_Salary.append(ms)
Minmax_Salary[0:20]
```

```python
Minimum_salary=Minmax_Salary[::2]
```

```python
Minimum_salary
```

```python
Maximum_salary=Minmax_Salary[1::2]
```

```python
Maximum_salary
```

```python
average_tags=driver.find_elements_by_xpath("//p[@class='averageCtc']")
for i in average_tags:
    ms=i.text
    Average_salary.append(ms)
Average_salary[0:10]
```

```python
experience_tags=driver.find_elements_by_xpath("//div[@class='salaries sbold-list-header']")
for i in experience_tags:
    ex=i.text
    Experience_required.append(ex)
Experience_required[0:10]
```

```python
print(len(Company_Name),len(Salary_Records),len(Average_salary),len(Minimum_salary),len(Maximum_salary),len(Experience_required))
```

```python
jobs=pd.DataFrame({})
jobs['Company_Name']=Company_Name
jobs['Salary_Records']=Salary_Records
jobs['Average_salary']=Average_salary
jobs['Minimum_salary']=Minimum_salary
jobs['Maximum_salary']=Maximum_salary
jobs['Experience_required']=Experience_required
```
